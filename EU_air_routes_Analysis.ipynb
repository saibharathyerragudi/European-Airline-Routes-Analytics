# Reading Raw data to Notebook


```python
from pyspark.sql import functions as F

# Start from the RAW table that was created from the CSV
raw_df = spark.table("workspace.eu_air_routes.europe_air_routes_raw")

display(raw_df.limit(5))
raw_df.printSchema()

```

# Data Cleaning,Preparation and analysis
## Clean “no” / invalid numeric values BEFORE casting

```python
df = raw_df

numeric_cols = [
    "id",
    "price",
    "common_duration",
    "flights_per_week",
    "arrival_airport_no_routes",
    "departure_latitude", "departure_longitude", "departure_altitude",
    "arrival_latitude", "arrival_longitude", "arrival_altitude"
]

for c in numeric_cols:
    df = df.withColumn(c, F.col(c).cast("double"))


```

#Feature Engineering

```python
from functools import reduce

day_cols = ["day1","day2","day3","day4","day5","day6","day7"]

# make _flag columns: 1 if "yes" (case-insensitive), else 0
for d in day_cols:
    df = df.withColumn(
        d + "_flag",
        F.when(F.lower(F.col(d)) == "yes", F.lit(1)).otherwise(F.lit(0))
    )

flag_cols = [d + "_flag" for d in day_cols]

# days_operated = number of days where dayX == "yes"
df = df.withColumn(
    "days_operated",
    reduce(lambda a, b: a + b, [F.col(c) for c in flag_cols])
)

# is_daily_route = flies all 7 days
df = df.withColumn(
    "is_daily_route",
    (F.col("days_operated") == F.lit(7))
)

```

## numeric version of flights_per_day

```python
# Example mapping for flights_per_day
df = df.withColumn(
    "flights_per_day_num",
    F.when(F.col("flights_per_day") == "0", 0.0)
     .when(F.col("flights_per_day") == "1 flight", 1.0)
     .when(F.col("flights_per_day") == "2 flights", 2.0)
     # handle ranges like "0-1 flights", "3-4 flights", etc.
     .when(F.col("flights_per_day").rlike("^[0-9]+-[0-9]+"),
           (
               F.split(F.col("flights_per_day"), "-").getItem(0).cast("double") +
               F.regexp_extract(F.col("flights_per_day"), "([0-9]+)\\s*flights?", 1).cast("double")
           ) / 2.0
     )
     .otherwise(None)
)
```

##  Distance(Haversine)

```python
from pyspark.sql.functions import radians, sin, cos, asin, sqrt

R = 6371.0  # Earth radius in km

df = df.withColumn("distance_km",
    2 * R * asin(
        sqrt(
            sin((radians(F.col("arrival_latitude") - F.col("departure_latitude")))/2) ** 2 +
            cos(radians(F.col("departure_latitude"))) *
            cos(radians(F.col("arrival_latitude"))) *
            sin((radians(F.col("arrival_longitude") - F.col("departure_longitude")))/2) ** 2
        )
    )
)
```

## Price per KM

```python
df = df.withColumn(
    "price_per_km",
    F.when(F.col("distance_km") > 0, F.col("price") / F.col("distance_km"))
)
```

## saving cleaned data as "eu_air_routes_clean"

```python
df.write.format("delta").mode("overwrite").saveAsTable(
    "workspace.eu_air_routes.eu_air_routes_clean"
)

clean_df = spark.table("workspace.eu_air_routes.eu_air_routes_clean")
display(clean_df.limit(5))
clean_df.printSchema()
```
