# Notebook Analysis 
## loading the cleaned data

```python
clean_df = spark.table("workspace.eu_air_routes.eu_air_routes_clean")
```

## Basic Schema Overview


```python
clean_df.printSchema()
```

## row count


```python
clean_df.count()
```

## Preview the Data

```python
display(clean_df.limit(10))
```

## statisticts for numeric columns

```python
numeric_cols = [
    "price", "common_duration", "flights_per_week",
    "arrival_airport_no_routes",
    "departure_latitude", "departure_longitude",
    "arrival_latitude", "arrival_longitude",
    "distance_km", "price_per_km", "days_operated"
]

clean_df.select(numeric_cols).describe().show()
```

## Count null values per column

```python
from pyspark.sql import functions as F

null_counts = clean_df.select([
    F.sum(F.col(c).isNull().cast("int")).alias(c + "_nulls") 
    for c in clean_df.columns
])

display(null_counts)
```

## Top 10 Busiest Departure Airports

```python
(
    clean_df.groupBy("iata_from")
    .agg(F.sum("flights_per_week").alias("total_flights"))
    .orderBy(F.col("total_flights").desc())
    .limit(10)
).show()
```

## distrubution of days operated

```python
clean_df.groupBy("days_operated").count().orderBy("days_operated").show()
```

## Basic correlation check (price vs distance)

```python
clean_df.select("price", "distance_km").stat.corr("price", "distance_km")
```

# Departure vs Arrival Cooridinates

```python
import matplotlib.pyplot as plt

geo_pdf = clean_df.select(
    "departure_latitude", "departure_longitude",
    "arrival_latitude", "arrival_longitude"
).dropna().limit(8000).toPandas()

plt.figure(figsize=(10, 10))

plt.scatter(
    geo_pdf["departure_longitude"],
    geo_pdf["departure_latitude"],
    alpha=0.35,
    label="Departure Points"
)

plt.scatter(
    geo_pdf["arrival_longitude"],
    geo_pdf["arrival_latitude"],
    alpha=0.35,
    label="Arrival Points"
)

plt.title("European Air Routes – Geographic Scatter of Departure vs Arrival Airports")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend()
plt.show()
```

#Distribution of Route Distances

```python
dist_pdf = clean_df.select("distance_km").dropna().toPandas()

plt.figure(figsize=(10, 5))
plt.hist(dist_pdf["distance_km"], bins=40, edgecolor="black")
plt.title("Distribution of Route Distances (km)")
plt.xlabel("Distance (km)")
plt.ylabel("Number of Routes")
plt.show()
```

# price vs distance relationship

```python
price_pdf = clean_df.select("distance_km", "price").dropna().toPandas()

plt.figure(figsize=(10, 6))
plt.scatter(price_pdf["distance_km"], price_pdf["price"], alpha=0.4)
plt.title("Relationship Between Flight Distance and Ticket Price")
plt.xlabel("Distance (km)")
plt.ylabel("Price (€)")
plt.show()
```

#Weekly Flight Volume by Departure Country

```python
country_df = (
    clean_df.groupBy("departure_country")
    .agg(F.sum("flights_per_week").alias("total_flights"))
    .orderBy(F.col("total_flights").desc())
).toPandas()

plt.figure(figsize=(14, 6))
plt.bar(country_df["departure_country"], country_df["total_flights"])
plt.xticks(rotation=45, ha="right")
plt.xlabel("Departure Country")
plt.ylabel("Total Weekly Flights")
plt.title("Total Weekly Flights by Departure Country")
plt.show()
```
